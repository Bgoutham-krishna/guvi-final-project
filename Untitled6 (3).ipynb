```python
fig, ax = plt.subplots(ncols=4, nrows=3, figsize=(30, 10))
ax = ax.flatten()
index = 0
for i in data.columns:
  if i != 'quality':
    sns.boxplot(y=i, data=data, ax=ax[index])
    index +=1
plt.tight_layout(pad=1)
plt.show()
```

There are many outliers but as they determine the quality of the wine we choose not to omit them.

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(ncols=2, nrows=5, figsize=(30, 10))
ax = ax.flatten()
index = 0

for i in data.columns:
    if i != 'quality':
        sns.barplot(x='quality', y=i, data=data, ax=ax[index])
        index += 1
        if index >= len(ax):
            break  # Exit the loop if index goes beyond the length of ax

plt.tight_layout(pad=1)
plt.show()
```

Based on the figures, Quality cannot be determined using Fixed acidity, Residual sugar, free sulfur dioxide and total sulfur dioxide as there is no significant trend seen.

Quality increases with -decrease in volatile acidity. -increase in citric acid. -decrease in chlorides. -decrease in pH. -increase in sulphates. -increase in alcohol.

```python
bins = (2, 6, 8)
group_names = ['Bad', 'Good']
data['quality'] = pd.cut(data['quality'], bins = bins, labels = group_names)
data.head()
```

From the above code have divided the quality of wine in two buckets:

Bad wine : range 2 - 6
Good wine : range 6 - 8


Now we will map the values of bad and good to 0 and 1 respectively, as machine learning models can perform calculation only on numerical data.

```python
print(data['quality'].unique())
```

```python
data['quality'] = data['quality'].map({'Bad' : 0, 'Good' : 1})
data.head(10)
```

```python
print(data['quality'].value_counts())
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(15, 5))
sns.countplot(data['quality'])
plt.show()
```

```python
X = data.iloc[:, :-1]
Y = (data.iloc[:, 11])
```

```python
print (Y)
```

```python
X_train, X_test, Y_train, Y_test = tts(X, Y, test_size=0.20, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler as ss
SS = ss()
X_train = SS.fit_transform(X_train)
X_test = SS.transform(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
rfc = RFC(n_estimators=25, criterion='gini', random_state=0,)
rfc.fit(X_train, Y_train)
Y_pred_rfc = rfc.predict(X_test)
print(classification_report(Y_test, Y_pred_rfc))
```

```python
print('\nConfussion matrix:')
print(CM(Y_test, Y_pred_rfc))
```

```python
knn = KNN(n_neighbors=2, metric='minkowski', p=2,)
knn.fit(X_train, Y_train)
Y_pred_knn = knn.predict(X_test)
print(classification_report(Y_test, Y_pred_knn))
print('\nConfussion matrix:')
print(CM(Y_test, Y_pred_knn))
```

```python
svc = SVC(kernel='rbf', gamma='scale', random_state=0)
svc.fit(X_train, Y_train)
Y_pred_svc = svc.predict(X_test)
print(classification_report(Y_test, Y_pred_svc))
print('\nConfussion matrix:')
print(CM(Y_test, Y_pred_svc))
```

```python
logisticRegression = LogisticRegression(solver='lbfgs', random_state=0)
logisticRegression.fit(X_train, Y_train)
Y_pred_logisticRegression = logisticRegression.predict(X_test)
print(classification_report(Y_test, Y_pred_logisticRegression))
print('\nConfussion matrix:')
print(CM(Y_test, Y_pred_logisticRegression))
```
